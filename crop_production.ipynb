```python
# Asking the user for the column name
column_name = input("Enter the column name for IQR calculation: ")

# Finding the IQR
Q1 = df3[column_name].quantile(0.25)
Q3 = df3[column_name].quantile(0.75)
IQR = Q3 - Q1

# Finding upper and lower limits
upper_limit = Q3 + (1.5 * IQR)
lower_limit = Q1 - (1.5 * IQR)

# Count the number of outliers BEFORE capping
num_outliers = ((df3[column_name] > upper_limit) | (df3[column_name] < lower_limit)).sum()

# Capping the outliers directly in df3 (no need to create df4)
df3.loc[df3[column_name] > upper_limit, column_name] = upper_limit
df3.loc[df3[column_name] < lower_limit, column_name] = lower_limit

# Displaying the number of modified values
print(f"Number of outliers capped in '{column_name}': {num_outliers}")

# Display the first few rows of the updated dataset
print(df3.head())
```

```python
#1.Top 7 RICE PRODUCTION State Data(Bar_plot)
top7 = df3.groupby("State Name")["RICE PRODUCTION (1000 tons)"].sum().nlargest(7)
ax=top7.plot(kind="bar",color="gold")
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x/100000:.1f}L'))
plt.title("Top 7 Rice Production States",color="green")
plt.xlabel("State Name")
plt.ylabel("Rice Production in Tons")
plt.show()
```

```python
#2.Top 5 Wheat Producing States Data(Bar_chart)

top5wheetbar=df3.groupby("State Name")["WHEAT PRODUCTION (1000 tons)"].sum().nlargest(5)
ax=top5wheetbar.plot(kind="bar",color="green")
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x/100000:.1f}L'))
plt.title("Top 5 Wheat Production States")
plt.xlabel("State Name")
plt.ylabel("Wheat Production in Tons")
```

```python
#3.Top 5 Wheat Producing States Data its percentage(%)(Pie_chart)

top5wheatpie = df3.groupby("State Name")["WHEAT PRODUCTION (1000 tons)"].sum().nlargest(5)
plt.figure(figsize=(8, 8))
plt.pie(
    top5wheatpie, 
    labels=top5wheatpie.index,  
    autopct='%1.1f%%', 
    startangle=140,
    colors=["gold", "lightblue", "lightgreen", "orange", "pink"],  
    wedgeprops={'edgecolor': 'black'} 
)

plt.title("Top 5 Wheat Producing States and Their Percentage Share")
plt.show()
```
